```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a demo dataset
np.random.seed(42)
demo_tips = pd.DataFrame({
    'tip': np.random.uniform(1, 10, 100)
})

# Plot KDE with fill
plt.figure(figsize=(10, 6))
sns.kdeplot(data=demo_tips['tip'], fill=True, color='mediumseagreen')

plt.title("KDE Plot of Tips (Demo Data)")
plt.xlabel("Tip Amount")
plt.ylabel("Density")
plt.grid(True)
plt.tight_layout()

# Save as PNG
plt.savefig("kdeplot_demo.png", dpi=300)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create demo dataset
np.random.seed(42)
demo_tips_box = pd.DataFrame({
    'total_bill': np.random.uniform(10, 50, 100),
    'day': np.random.choice(['Thur', 'Fri', 'Sat', 'Sun'], size=100)
})

# Generate box plot
plt.figure(figsize=(10, 6))
sns.boxplot(
    x='day',
    y='total_bill',
    data=demo_tips_box,
    palette='Pastel1'
)

plt.title("Box Plot of Total Bill by Day (Demo Data)")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.grid(True, axis='y')
plt.tight_layout()

# Save for PPT
plt.savefig("boxplot_demo.png", dpi=300)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create demo dataset
np.random.seed(42)
demo_tips_pair = pd.DataFrame({
    'total_bill': np.random.uniform(10, 50, 100),
    'tip': np.random.uniform(1, 10, 100),
    'size': np.random.randint(1, 6, 100),
    'sex': np.random.choice(['Male', 'Female'], 100)
})

# Generate pair plot
pairplot = sns.pairplot(demo_tips_pair, hue='sex', kind='scatter', palette='Set2')
pairplot.fig.suptitle("Pair Plot (Demo Data)", y=1.02)

# Save for PPT
pairplot.savefig("pairplot_demo.png", dpi=300)
plt.show()
```
