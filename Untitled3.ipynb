```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.title("My Graph")
plt.xlabel("Hello")
plt.ylabel("Hello")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y , "o" , linestyle=('dashed') , color = 'hotpink')
plt.title("My Graph")
plt.xlabel("Hello")
plt.ylabel("Hello")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]


plt.plot(x,y1 , linestyle=('dashed') , color = 'blue')
plt.plot(x,y2 , linestyle=('dashed') , color = 'blue')
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]


plt.plot(x,y1 , linestyle=('dashed') , color = 'yellow')
plt.plot(x,y2 , linestyle=('dashed') , color = 'green')


plt.grid()
plt.title("My Graph")
plt.xlabel("Hello")
plt.ylabel("heyyy")
plt.legend(["hehehehhe😁"])

plt.show()
```
